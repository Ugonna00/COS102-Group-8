```python
import tkinter as tk 
from tkinter import ttk
from PIL import Image, ImageTk
import psycopg2
from psycopg2 import sql

conn = psycopg2.connect(
    host="localhost",
    database="ewaste_db",
    user="postgres",
    password="cos101",
    port="5433"
)
root = tk.Tk()
root.title("record")
root.geometry("600x400")

Img = Image.open("rec.png")
Img = Img.resize((150, 150))
pic = ImageTk.PhotoImage(Img)

img = Image.open("rec.png")
img = img.resize((90, 90))
photo = ImageTk.PhotoImage(img)

left_frame = tk.Frame(root,bg="green",width=70,height=400)
left_frame.place(x=50,y=0)

image_label = tk.Label(root,image=photo)
image_label.place(x=135, y=30)

name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
name_label.place(x=230, y=30)
text1_label = tk.Label(root, text="Online Waste Tracker", font=("Arial", 18, "bold"))
text1_label.place(x=230, y=83)


table = ttk.Treeview(
    root,
    columns=("ID","USERNAME", "ITEM", "QTY"),
    show="headings"
)

table.heading("ID", text="ID")
table.heading("USERNAME", text="USERNAME")
table.heading("ITEM", text="ITEM")
table.heading("QTY", text="QTY")


table.column("ID", width=100)
table.column("USERNAME", width=100)
table.column("ITEM", width=120)
table.column("QTY", width=100)

cursor = conn.cursor()
cursor.execute("SELECT id, username, item, quantity FROM waste_logs")

records = cursor.fetchall()

for row in records:
    table.insert("", tk.END, values=row)


table.place(x=150,y=140)


root.mainloop()
```
