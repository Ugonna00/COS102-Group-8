```python
import tkinter as tk
from tkinter import messagebox as msgbox
from PIL import Image, ImageTk


root = tk.Tk()
root.title("Eco tips")
root.geometry("600x400")

Img = Image.open("rec.png")
Img = Img.resize((150, 150))
pic = ImageTk.PhotoImage(Img)

img = Image.open("rec.png")
img = img.resize((90, 90))
photo = ImageTk.PhotoImage(img)

left_frame = tk.Frame(root,bg="green",width=70,height=400)
left_frame.place(x=50,y=0)

image_label = tk.Label(root,image=photo)
image_label.place(x=135, y=30)

name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
name_label.place(x=230, y=30)
text1_label = tk.Label(root, text="Online Waste Tracker", font=("Arial", 18, "bold"))
text1_label.place(x=230, y=83)
canvas = tk.Canvas(root, width=20, height=310)
canvas.place(x=210, y=110)
canvas.create_line(10, 10, 10, 285, fill="green", width=1)

def on_closing():
    if msgbox.askyesno("Quit", "Are you sure you want to exit?"):
        root.destroy()

root.protocol("WM_DELETE_WINDOW", on_closing)

tk.Label(text="Eco Tips & Awareness",font=("Arial", 11, "bold")).place(x=231,y=120)
card = tk.Frame()
card.place(x=230,y=145)
tk.Label(card, bd=1,relief="solid",padx=2,pady=2,
        text="Don't trash it, recycle it: \nOld phones, cables, and batteries contain toxic materials. \nAlways drop them at a certified e-waste point.",
        justify="left",highlightbackground="green").pack(pady=9)
tk.Label(card,bd=1,relief="solid",padx=5,pady=2,
        text="Battery safety:\nNever throw batteries in regular bins. They can leak acid \nor cause fires in landfills.",
        justify="left").pack(padx=9,pady=2)
tk.Label(card, bd=1,relief="solid",padx=1,pady=2,
        text="Phones have value:\nA single smartphone contains gold, silver, copper, and rare\nearth metals, which are all recoverable through recycling.",
        justify="left").pack(pady=9)
tk.Label(card, bd=1,relief="solid",padx=2,pady=2,
        text="Scale matters:\nThe world generates over 50 million tonnes of e-waste per\nyear. Only 20% is formally recycled.",
        justify="left").pack()

root.mainloop()
```
