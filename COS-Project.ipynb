```python
import tkinter as tk
from tkinter import messagebox as msgbox
from PIL import Image, ImageTk
import json
import os


class EcoTrackEnhancer:
    BG       = "#0a0f0a"
    ACCENT   = "#3ddc6a"
    ACCENT2  = "#27a34a"
    TEXT     = "#e0f0e0"
    MUTED    = "#5a7a5a"
    ENTRY_BG = "#111f11"
    QUOTE    = '"Every item recycled is armour against digital waste"'

    def __init__(self, root):
        self.root = root
        root.configure(bg=self.BG)

    def apply(self, widgets, show_quote=False, quote_y=310):
        styles = {
            "sidebars":  dict(bg=self.ACCENT2),
            "accent":    dict(bg=self.BG, fg=self.ACCENT),
            "muted":     dict(bg=self.BG, fg=self.MUTED),
            "entries":   dict(bg=self.ENTRY_BG, fg=self.TEXT, relief="flat",
                              insertbackground=self.ACCENT, highlightthickness=1,
                              highlightbackground=self.ACCENT2),
            "buttons":   dict(bg=self.ACCENT2, fg=self.BG, relief="flat", bd=0),
            "secondary": dict(bg=self.BG, fg=self.ACCENT, relief="flat", bd=1),
            "dividers":  dict(bg=self.BG, highlightthickness=0),
        }
        for key, cfg in styles.items():
            for w in widgets.get(key, []):
                w.configure(**cfg)
                if key == "dividers":
                    w.delete("all")
                    w.create_line(10, 10, 10, 170, fill=self.ACCENT2, width=1)
        self.root.configure(bg=self.BG)
        if show_quote:
            tk.Label(self.root, text=self.QUOTE, font=("Arial", 8, "italic"),
                     fg=self.ACCENT2, bg=self.BG, wraplength=500).place(x=80, y=quote_y)


class GoalTracker:
    def __init__(self):
        self.file = "ecotrack_goals.json"
        self.load()
    
    def load(self):
        if os.path.exists(self.file):
            with open(self.file, 'r') as f:
                d = json.load(f)
                self.daily, self.weekly, self.total, self.today = d.get('daily', 10), d.get('weekly', 50), d.get('total', 0), d.get('today', 0)
        else:
            self.daily, self.weekly, self.total, self.today = 10, 50, 0, 0
    
    def save(self):
        with open(self.file, 'w') as f:
            json.dump({'daily': self.daily, 'weekly': self.weekly, 'total': self.total, 'today': self.today}, f)
    
    def add(self, qty):
        self.today += qty
        self.total += qty
        self.save()
        msgs = []
        if self.today >= self.daily: msgs.append(f"🎉 Daily goal reached! ({self.today}/{self.daily})")
        if self.today >= self.weekly: msgs.append(f"🏆 Weekly goal reached! ({self.today}/{self.weekly})")
        return msgs
    
    def set_daily(self, amt): self.daily = amt; self.save()
    def set_weekly(self, amt): self.weekly = amt; self.save()


def clear_page():
    for widget in root.winfo_children():
        widget.destroy()
        
def signed_up():
    msgbox.showinfo("Info", "Signed up successfully!")
    login()  

def sign_up():
    clear_page()
    
    left_frame = tk.Frame(root,bg="green",width=70,height=400)
    left_frame.place(x=50,y=0)
    
    image_label = tk.Label(root,image=photo)
    image_label.place(x=135, y=30)
    
    name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
    name_label.place(x=230, y=30)
    text1_label = tk.Label(root, text="Online Waste Tracker", font=("Arial", 18, "bold"))
    text1_label.place(x=230, y=83)
    
    canvas = tk.Canvas(root, width=20, height=180)
    canvas.place(x=210, y=120)
    canvas.create_line(10, 10, 10, 170, fill="green", width=1)
    
    label = tk.Label(root,text="Set Username:",font=("Arial", 10))
    label.place(x=230, y=140)
    entry = tk.Entry(root,width=40)
    entry.place(x=234, y=160)

    pass_ = tk.Label(root,text="Set Password:",font=("Arial", 10))
    pass_.place(x=230, y=195)
    pass_entry = tk.Entry(root,width=40,show="*")
    pass_entry.place(x=234, y=215)

    log_button = tk.Button(root,text="Sign Up",bg="white",fg="green",width=34,command=signed_up)
    log_button.place(x=234, y=250)

    footer = tk.Label(root, text="EcoTrack © 2026",font=("Arial", 9),fg="gray")
    footer.place(x=280, y=370)

    enhancer.apply({
        "sidebars":  [left_frame],
        "accent":    [name_label],
        "muted":     [text1_label, label, pass_, footer],
        "entries":   [entry, pass_entry],
        "buttons":   [log_button],
        "dividers":  [canvas],
    })
    image_label.configure(bg=enhancer.BG)


def set_goals_dialog():
    d = tk.Toplevel(root)
    d.title("Set Goals")
    d.geometry("300x200")
    d.configure(bg=enhancer.BG)
    
    tk.Label(d, text="Daily Goal:", bg=enhancer.BG, fg=enhancer.TEXT).pack(pady=10)
    daily_e = tk.Entry(d)
    daily_e.insert(0, str(goal.daily))
    daily_e.pack()
    
    tk.Label(d, text="Weekly Goal:", bg=enhancer.BG, fg=enhancer.TEXT).pack(pady=10)
    weekly_e = tk.Entry(d)
    weekly_e.insert(0, str(goal.weekly))
    weekly_e.pack()
    
    def save():
        try:
            nd, nw = int(daily_e.get()), int(weekly_e.get())
            if nd > 0 and nw > 0:
                goal.set_daily(nd)
                goal.set_weekly(nw)
                msgbox.showinfo("Success", "Goals updated!")
                d.destroy()
                show_goals()
            else:
                msgbox.showwarning("Error", "Enter positive numbers!")
        except:
            msgbox.showwarning("Error", "Enter valid numbers!")
    
    tk.Button(d, text="Save", bg="green", fg="white", command=save).pack(pady=20)


def show_goals():
    clear_page()
    
    left_frame = tk.Frame(root,bg="green",width=70,height=400)
    left_frame.place(x=50,y=0)
    
    image_label = tk.Label(root,image=photo)
    image_label.place(x=135, y=30)
    
    name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
    name_label.place(x=230, y=30)
    tk.Label(root, text="Your Recycling Goals", font=("Arial", 18, "bold")).place(x=230, y=83)
    
    canvas = tk.Canvas(root, width=20, height=180)
    canvas.place(x=210, y=120)
    canvas.create_line(10, 10, 10, 170, fill="green", width=1)
    
    daily_pct = int((goal.today / goal.daily) * 100) if goal.daily > 0 else 0
    weekly_pct = int((goal.today / goal.weekly) * 100) if goal.weekly > 0 else 0
    
    # Daily goal
    tk.Label(root, text="📅 DAILY GOAL", font=("Arial", 12, "bold"), fg="green").place(x=260, y=140)
    tk.Label(root, text=f"Progress: {goal.today} / {goal.daily} items", font=("Arial", 10)).place(x=260, y=165)
    bar_bg = tk.Canvas(root, width=250, height=20, bg="#ddd", highlightthickness=1)
    bar_bg.place(x=260, y=190)
    w = int((goal.today / goal.daily) * 240) if goal.daily > 0 else 0
    tk.Canvas(root, width=w, height=18, bg="green", highlightthickness=0).place(x=261, y=191)
    tk.Label(root, text=f"{daily_pct}%", font=("Arial", 9, "bold"), fg="green").place(x=490, y=192)
    
    # Weekly goal
    tk.Label(root, text="📆 WEEKLY GOAL", font=("Arial", 12, "bold"), fg="green").place(x=260, y=225)
    tk.Label(root, text=f"Progress: {goal.today} / {goal.weekly} items", font=("Arial", 10)).place(x=260, y=250)
    bar_bg2 = tk.Canvas(root, width=250, height=20, bg="#ddd", highlightthickness=1)
    bar_bg2.place(x=260, y=275)
    w2 = int((goal.today / goal.weekly) * 240) if goal.weekly > 0 else 0
    tk.Canvas(root, width=w2, height=18, bg="green", highlightthickness=0).place(x=261, y=276)
    tk.Label(root, text=f"{weekly_pct}%", font=("Arial", 9, "bold"), fg="green").place(x=490, y=277)
    
    tk.Label(root, text=f"🌟 TOTAL: {goal.total} items", font=("Arial", 11, "bold"), fg="green").place(x=260, y=310)
    btn = tk.Button(root, text="✏️ SET NEW GOALS", bg="green", fg="white", width=20, command=set_goals_dialog)
    btn.place(x=330, y=340)
    tk.Label(root, text="EcoTrack © 2026", font=("Arial", 9), fg="gray").place(x=280, y=370)
    
    image_label.configure(bg=enhancer.BG)
    enhancer.apply({
        "sidebars": [left_frame],
        "accent": [name_label],
        "muted": [tk.Label()],
        "buttons": [btn],
        "dividers": [canvas],
    })


def log_item():
    clear_page()
    
    left_frame = tk.Frame(root,bg="green",width=70,height=400)
    left_frame.place(x=50,y=0)
    
    image_label = tk.Label(root,image=photo)
    image_label.place(x=135, y=30)
    
    name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
    name_label.place(x=230, y=30)
    text1_label = tk.Label(root, text="Online Waste Tracker", font=("Arial", 18, "bold"))
    text1_label.place(x=230, y=83)
    
    canvas = tk.Canvas(root, width=20, height=180)
    canvas.place(x=210, y=120)
    canvas.create_line(10, 10, 10, 170, fill="green", width=1)
    
    log_label = tk.Label(root,text="Item:",font=("Arial", 10))
    log_label.place(x=230, y=140)
    log_entry = tk.Entry(root,width=30)
    log_entry.place(x=234, y=160)

    quantity = tk.Label(root,text="Quantity:",font=("Arial", 10))
    quantity.place(x=230, y=195)
    quantity_entry = tk.Entry(root,width=30)
    quantity_entry.place(x=234, y=215)

    def log_and_update():
        try:
            qty = int(quantity_entry.get())
            if qty > 0:
                achievements = goal.add(qty)
                msg = f"Logged {qty} x {log_entry.get()}!\n"
                if achievements:
                    msg += "\n".join(achievements)
                msgbox.showinfo("Success!", msg)
                log_entry.delete(0, tk.END)
                quantity_entry.delete(0, tk.END)
            else:
                msgbox.showwarning("Error", "Please enter a positive quantity!")
        except ValueError:
            msgbox.showwarning("Error", "Please enter a valid quantity!")

    log_button = tk.Button(root,text="Log item",bg="green",fg="white",width=20,command=log_and_update)
    log_button.place(x=234, y=250)
    
    goals_button = tk.Button(root,text="🎯 VIEW GOALS",bg="green",fg="white",width=20,command=show_goals)
    goals_button.place(x=390, y=250)

    footer = tk.Label(root, text="EcoTrack © 2026",font=("Arial", 9),fg="gray")
    footer.place(x=280, y=370)

    enhancer.apply({
        "sidebars":  [left_frame],
        "accent":    [name_label],
        "muted":     [text1_label, log_label, quantity, footer],
        "entries":   [log_entry, quantity_entry],
        "buttons":   [log_button, goals_button],
        "dividers":  [canvas],
    })
    image_label.configure(bg=enhancer.BG)

        
def login():
    clear_page()
    
    left_frame = tk.Frame(root,bg="green",width=70,height=400)
    left_frame.place(x=50,y=0)
    
    image_label = tk.Label(root,image=photo)
    image_label.place(x=135, y=30)
    
    name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
    name_label.place(x=230, y=30)
    text1_label = tk.Label(root, text="Online Waste Tracker", font=("Arial", 18, "bold"))
    text1_label.place(x=230, y=83)
    
    canvas = tk.Canvas(root, width=20, height=180)
    canvas.place(x=210, y=120)
    canvas.create_line(10, 10, 10, 170, fill="green", width=1)
    
    username_label = tk.Label(root,text="Username:",font=("Arial", 10))
    username_label.place(x=230, y=140)
    username_entry = tk.Entry(root,width=40)
    username_entry.place(x=234, y=160)

    password_label = tk.Label(root,text="Password:",font=("Arial", 10))
    password_label.place(x=230, y=195)
    password_entry = tk.Entry(root,width=40,show="*")
    password_entry.place(x=234, y=215)

    login_button = tk.Button(root,text="LOGIN",bg="green",fg="white",width=14, command=log_item)
    login_button.place(x=234, y=250)
    signup_button = tk.Button(root,text="SIGN UP",width=14, command=sign_up, bg="white",fg="green")
    signup_button.place(x=370, y=250)

    footer = tk.Label(root, text="EcoTrack © 2026",font=("Arial", 9),fg="gray")
    footer.place(x=280, y=370)

    enhancer.apply({
        "sidebars":  [left_frame],
        "accent":    [name_label],
        "muted":     [text1_label, username_label, password_label, footer],
        "entries":   [username_entry, password_entry],
        "buttons":   [login_button],
        "secondary": [signup_button],
        "dividers":  [canvas],
    })
    image_label.configure(bg=enhancer.BG)


root = tk.Tk()
root.title("Welcome")
root.geometry("600x400")

enhancer = EcoTrackEnhancer(root)
goal = GoalTracker()  

Img = Image.open("rec.png")
Img = Img.resize((150, 150))
pic = ImageTk.PhotoImage(Img)

img = Image.open("rec.png")
img = img.resize((90, 90))
photo = ImageTk.PhotoImage(img)

left_frame = tk.Frame(root,bg="green",width=70,height=400)
left_frame.place(x=50,y=0)

image_label = tk.Label(root,image=photo)
image_label.place(x=135, y=30)

name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
name_label.place(x=230, y=30)
text1_label = tk.Label(root, text="Online Waste Tracker", font=("Arial", 18, "bold"))
text1_label.place(x=230, y=83)

canvas = tk.Canvas(root, width=20, height=120)
canvas.place(x=210, y=120)
canvas.create_line(10, 10, 10, 110, fill="green", width=1)

text2_label = tk.Label(root, text="Welcome to ECO Track!", font=("Arial", 15))
text2_label.place(x=230, y=135)
text3_label = tk.Label(root, text="Log recycled electronic parts online", font=("Arial", 15))
text3_label.place(x=230, y=160)
text4_label = tk.Label(root, text="Contribute to a cleaner tommorow", font=("Arial", 15))
text4_label.place(x=230, y=190)

continue_ = tk.Button(root, text="Continue", bg="green", fg="white", pady=5, width=20, command=login)
continue_.place(x=350, y=250)

footer = tk.Label(root, text="EcoTrack © 2026",font=("Arial", 9),fg="gray")
footer.place(x=280, y=370)

enhancer.apply({
    "sidebars": [left_frame],
    "accent":   [name_label],
    "muted":    [text1_label, text2_label, text3_label, text4_label, footer],
    "buttons":  [continue_],
    "dividers": [canvas],
}, show_quote=True, quote_y=305)
image_label.configure(bg=enhancer.BG)

root.mainloop()
```
