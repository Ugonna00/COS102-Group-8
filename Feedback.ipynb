```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

root = tk.Tk()
root.title("feedback")
root.geometry("600x400")

Img = Image.open("rec.png")
Img = Img.resize((150, 150))
pic = ImageTk.PhotoImage(Img)

img = Image.open("rec.png")
img = img.resize((90, 90))
photo = ImageTk.PhotoImage(img)

left_frame = tk.Frame(root,bg="green",width=70,height=400)
left_frame.place(x=50,y=0)

image_label = tk.Label(root,image=photo)
image_label.place(x=135, y=30)

name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
name_label.place(x=230, y=30)
text1_label = tk.Label(root, text="Online Waste Tracker", font=("Arial", 18, "bold"))
text1_label.place(x=230, y=83)

canvas = tk.Canvas(root, width=20, height=180)
canvas.place(x=210, y=120)
canvas.create_line(10, 10, 10, 170, fill="green", width=1)

def on_closing():
    if messagebox.askyesno("Quit", "Are you sure you want to exit?"):
        root.destroy()

root.protocol("WM_DELETE_WINDOW", on_closing)

def pop_up():
    messagebox.showinfo("Successful", f"Thanks for your feedback!")
    feedback.delete("1.0", tk.END)  

card = tk.Frame(root)
card.place(x=230, y=125, width=300, height=260)

feedback_label = tk.Label(card, font=('Arial', 14, 'bold'), text='How is the app?')
feedback_label.place(x=0)

sub_label = tk.Label(card, font=('Arial', 10), text='We value your feedback.')
sub_label.place(x=0,y=24)


feedback = tk.Text(card, font=('Arial', 10), height=7, bd=1, relief="solid")
feedback.place(x=0,y=45)

submit_button = tk.Button(root,text="SUBMIT",bg="green",fg="white",width=14, command=pop_up)
submit_button.place(x=422, y=300)

root.mainloop()
```
