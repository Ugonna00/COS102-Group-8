```python
import psycopg2

# =========================
# DATABASE CONNECTION
# =========================

def connect_db():
    conn = psycopg2.connect(
        host="localhost",
        database="ecotrack",
        user="postgres",
        password="yourpassword",
        port="5432"
    )
    cursor = conn.cursor()
    return conn, cursor


# =========================
# CREATE TABLES
# =========================

def create_tables():
    conn, cursor = connect_db()

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS users(
        id SERIAL PRIMARY KEY,
        username VARCHAR(100) UNIQUE NOT NULL,
        password VARCHAR(100) NOT NULL
    )
    """)

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS waste_logs(
        id SERIAL PRIMARY KEY,
        item VARCHAR(100) NOT NULL,
        quantity INTEGER NOT NULL
    )
    """)

    conn.commit()
    conn.close()


# =========================
# SIGN UP USER
# =========================

def save_user(username, password):
    conn, cursor = connect_db()

    cursor.execute("""
        INSERT INTO users(username, password)
        VALUES (%s, %s)
    """, (username, password))

    conn.commit()
    conn.close()


# =========================
# LOGIN CHECK
# =========================

def check_login(username, password):
    conn, cursor = connect_db()

    cursor.execute("""
        SELECT * FROM users
        WHERE username=%s AND password=%s
    """, (username, password))

    user = cursor.fetchone()

    conn.close()
    return user


# =========================
# SAVE WASTE LOG
# =========================

def save_log(item, quantity):
    conn, cursor = connect_db()

    cursor.execute("""
        INSERT INTO waste_logs(item, quantity)
        VALUES (%s, %s)
    """, (item, quantity))

    conn.commit()
    conn.close()
```
