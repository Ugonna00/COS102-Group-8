```python
import tkinter as tk
from tkinter import messagebox
import csv
import os

root = tk.Tk()
root.title("ECO Track")
root.geometry("600x400")
root.resizable(False, False)

current_user = ""

def clear_page():
    for widget in root.winfo_children():
        widget.destroy()

def save_to_csv(item_name, qty):
    file_exists = os.path.isfile("waste_log.csv")
    
    with open("waste_log.csv", "a", newline="") as file:
        writer = csv.writer(file)
        
        if not file_exists:
            writer.writerow(["Username", "Item", "Quantity"])
        
        writer.writerow([current_user, item_name, qty])

def sign_up():
    clear_page()
    
    left_frame = tk.Frame(root, bg="green", width=70, height=400)
    left_frame.place(x=50, y=0)
    
    name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
    name_label.place(x=230, y=30)
    
    label = tk.Label(root, text="Set Username:", font=("Arial", 10))
    label.place(x=230, y=140)
    entry = tk.Entry(root, width=40)
    entry.place(x=234, y=160)

    pass_ = tk.Label(root, text="Set Password:", font=("Arial", 10))
    pass_.place(x=230, y=195)
    pass_entry = tk.Entry(root, width=40, show="*")
    pass_entry.place(x=234, y=215)

    def signed_up():
        user = entry.get().strip()
        pw = pass_entry.get().strip()
        
        if user == "" or pw == "":
            messagebox.showwarning("Missing Info", "Username and Password cannot be empty")
            return
        
        messagebox.showinfo("Success", "Account created! Please log in.")
        login()

    log_button = tk.Button(root, text="Sign Up", bg="white", fg="green", width=34, command=signed_up)
    log_button.place(x=234, y=250)

def log_item():
    clear_page()
    
    left_frame = tk.Frame(root, bg="green", width=70, height=400)
    left_frame.place(x=50, y=0)
    
    name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
    name_label.place(x=230, y=30)
    text1_label = tk.Label(root, text="Online Waste Tracker", font=("Arial", 18, "bold"))
    text1_label.place(x=230, y=83)
    
    welcome = tk.Label(root, text="Welcome, " + current_user, font=("Arial", 12))
    welcome.place(x=230, y=115)

    log_label = tk.Label(root, text="Item:", font=("Arial", 10))
    log_label.place(x=230, y=140)
    log_entry = tk.Entry(root, width=40)
    log_entry.place(x=234, y=160)

    quantity = tk.Label(root, text="Quantity:", font=("Arial", 10))
    quantity.place(x=230, y=195)
    quantity_entry = tk.Entry(root, width=40)
    quantity_entry.place(x=234, y=215)

    def submit_item():
        item = log_entry.get().strip()
        qty = quantity_entry.get().strip()
        
        if item == "" or qty == "":
            messagebox.showwarning("Missing Info", "Item and Quantity cannot be empty")
            return
        
        try:
            float(qty)
        except ValueError:
            messagebox.showerror("Invalid Input", "Quantity must be a number")
            return
        
        save_to_csv(item, qty)
        messagebox.showinfo("Saved", "Item logged successfully!")
        
        log_entry.delete(0, tk.END)
        quantity_entry.delete(0, tk.END)

    log_button = tk.Button(root, text="Log item", bg="green", fg="white", width=34, command=submit_item)
    log_button.place(x=234, y=250)
    
    back_button = tk.Button(root, text="Log Out", width=10, command=login)
    back_button.place(x=500, y=10)

def login():
    clear_page()
    
    left_frame = tk.Frame(root, bg="green", width=70, height=400)
    left_frame.place(x=50, y=0)
    
    name_label = tk.Label(root, text="ECO Track", fg="green", font=("Arial", 37, "bold"))
    name_label.place(x=230, y=30)
    text1_label = tk.Label(root, text="Online Waste Tracker", font=("Arial", 18, "bold"))
    text1_label.place(x=230, y=83)
    
    canvas = tk.Canvas(root, width=20, height=180)
    canvas.place(x=210, y=120)
    canvas.create_line(10, 10, 10, 170, fill="green", width=1)
    
    username_label = tk.Label(root, text="Username:", font=("Arial", 10))
    username_label.place(x=230, y=140)
    username_entry = tk.Entry(root, width=40)
    username_entry.place(x=234, y=160)

    password_label = tk.Label(root, text="Password:", font=("Arial", 10))
    password_label.place(x=230, y=195)
    password_entry = tk.Entry(root, width=40, show="*")
    password_entry.place(x=234, y=215)

    def try_login():
        global current_user
        user = username_entry.get().strip()
        pw = password_entry.get().strip()
        
        if user == "" or pw == "":
            messagebox.showwarning("Missing Info", "Username and Password cannot be empty")
            return
        
        current_user = user
        log_item()

    login_button = tk.Button(root, text="LOGIN", bg="green", fg="white", width=14, command=try_login)
    login_button.place(x=234, y=250)
    signup_button = tk.Button(root, text="SIGN UP", width=14, command=sign_up, bg="white", fg="green")
    signup_button.place(x=370, y=250)

# Start app
login()
root.mainloop()
```
